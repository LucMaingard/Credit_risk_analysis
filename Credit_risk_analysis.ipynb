```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

data = pd.read_csv('german_credit.csv')
data.head()
```

```python
#check all columns
data.dtypes
```

```python
#find missing values
missing = data.isnull().sum()
missing
```

```python
#check unique values in columns with missing
data['Creditability'].unique()
```

```python
#define test and get val data
feature_cols = data.columns.drop('Creditability')
y = data['Creditability']

# train test data
train_full = pd.read_csv('Training50.csv')
test_full = pd.read_csv('Test50.csv')

#validate data
drop_cols = ['Creditability', 'Unnamed: 0']
val_X = train_full.drop(drop_cols, axis=1)
test_X = test_full.drop(drop_cols, axis=1)
val_y = train_full['Creditability']
test_y = test_full['Creditability']

#train data
train_X = data[feature_cols]
train_y = y
```

```python
val_X.head()
```

```python
#Define and fit model
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(train_X, train_y)
```

```python
#predict
preds = model.predict(val_X)
```

```python
#score
from sklearn.metrics import mean_absolute_error

scores = mean_absolute_error(val_y, preds)
print('MAE Score: ', scores)
```
